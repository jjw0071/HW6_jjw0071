```python
import csv
import matplotlib.pyplot as plt

def main():
    f = open('result.csv')
    data = csv.reader(f)
    result = []
    i = 0
    
    plt.rcParams["figure.figsize"] = [11,7]
    fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,sharex=True)    
    fig.suptitle('Results of randomly thrown dice')
    for row in data:
        result.append(int(row[0]))
        i=i+1
        
        if i == 100:
            ax1.hist(result)
            #ax1.ylabel('times')
            result = []
            
        if i == 1100:            
            ax2.hist(result)
            #ax2.ylabel('times')
            result = []
            
            
        if i == 11100:
            ax3.hist(result)
            #ax3.ylabel('times')
            result = []
            
            
        if i == 111100:
            ax4.hist(result)
            #ax4.ylabel('times')
            result = []
            
            
    f.close()
    plt.xticks([1,2,3,4,5,6])
    plt.xlabel('result')
    plt.ylabel('times')
    plt.show()
    print("""랜덤하게 주사위를 던진 결과를 보면 100번만 시행했을 때는 각 숫자가 나온 횟수가 유사하지 않고 차이가 큰 모습을 보이지만, 횟수를 증가하여 1000번, 10000번, 100000번 시행했을 때의 결과들을 살펴보면 횟수가 증가할 수록 점점 각 숫자가 나온 횟수가 유사해지는 모습을 확인할 수 있다.""")
    
    
    
if __name__ == '__main__':
    main()
```
