```python
import csv
import matplotlib.pyplot as plt

def main():
    plt.rcParams['font.family'] ='Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] =False
    f = open('jejupopulation.csv')
    data = csv.reader(f)
    next(data)
    next(data)
    year = []
    for i in range(2013,2023):
        year.append(i)
        
    male = []
    female = []   
    
    plt.title('제주특별자치도 성별 인구수')
    plt.xlabel('연도')
    plt.ylabel('인구수')
    plt.xlim((2012,2023))

    
    for row in data :
        if '제주특별자치도  ' in row[1]:
            for i in row[4::4]:
                male.append(int(i))
            
            for i in row[5::4]:
                female.append(int(i))
    
    plt.bar(year,male, label = 'male')
    plt.bar(year,female, color = 'orange' , label = 'female')
    plt.xticks(year)
    plt.legend()    
    f.close()
    plt.show()
    
    print("해당 그래프는 제주특별자치도 전체의 인구 수를 성별로 비교하여 겹쳐 놓은 그래프이다.")
    print("파란색은 연도별 남성의 수이고, 주황색은 여성의 수 이다.")
    print("파란색막대의 길이가 주황색 막대의 길이보다 길어 위로 파란색이 보이는 모습을 확인할 수 있다.")
    print("그 결과 2013년부터 2022년까지 10년 내내 남성의 수가 더 많았다는 사실을 확인할 수 있다.")
    print("따라서 제주도에 여성의 비율이 높다는 속설은 옳지 않은 속설이다.")
    
    
    
if __name__ == '__main__':
    main() 
```
